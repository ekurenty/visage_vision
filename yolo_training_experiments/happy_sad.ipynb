```python
import os
import shutil
from google.colab import drive
drive.mount('/content/drive')
```

```python
# copying happy and sad images and label to new folders
# created variables for the sources and desinanation folders

img_train_path = '/content/drive/MyDrive/facial_expressions_dataset/new_yolo_images/images/train'
img_val_path = '/content/drive/MyDrive/facial_expressions_dataset/new_yolo_images/images/val'

label_train_path = '/content/drive/MyDrive/facial_expressions_dataset/new_yolo_images/labels/train'
label_val_path = '/content/drive/MyDrive/facial_expressions_dataset/new_yolo_images/labels/val'

dest_img_train_path = '/content/drive/MyDrive/facial_expressions_dataset/hs_dataset/images/train'
dest_img_val_path = '/content/drive/MyDrive/facial_expressions_dataset/hs_dataset/images/val'

dest_lab_train_path = '/content/drive/MyDrive/facial_expressions_dataset/hs_dataset/labels/train'
dest_lab_val_path = '/content/drive/MyDrive/facial_expressions_dataset/hs_dataset/labels/val'
```

```python
for file_name in os.listdir(img_train_path):
   if file_name.startswith('happy') or file_name.startswith('sad'):
     shutil.copy(f'{img_train_path}/{file_name}', dest_img_train_path)
```

```python
for file_name in os.listdir(img_val_path):
  if file_name.startswith('happy') or file_name.startswith('sad'):
    shutil.copy(f'{img_val_path}/{file_name}', dest_img_val_path)
```

```python
for file_name in os.listdir(label_train_path):
  if file_name.startswith('happy') or file_name.startswith('sad'):
    shutil.copy(f'{label_train_path}/{file_name}', dest_lab_train_path)
```

```python
for file_name in os.listdir(label_val_path):
  if file_name.startswith('happy') or file_name.startswith('sad'):
    shutil.copy(f'{label_val_path}/{file_name}', dest_lab_val_path)
```

```python
# next step is to overwrite the annotation text files for happy and sad since the labels will now by 0 and 1 instead of 4 and 6

class_dictionary = {
    'happy': 0,
    'sad': 1,
 }
```

```python
# reannotating val files

test_path = dest_lab_val_path

for filename in os.listdir(test_path):
    filename = "/".join([test_path, filename])

    for key, value in class_dictionary.items():
        if key in filename:
            line = ""

            with open(filename, 'r') as file:
                line = file.readline()

            line = f"{value} 0.500000 0.500000 1.000000 1.000000"

            with open(filename, 'w') as file:
                file.write(line)
```

```python
# reannotating train files

test_path = dest_lab_train_path

for filename in os.listdir(test_path):
    filename = "/".join([test_path, filename])

    for key, value in class_dictionary.items():
        if key in filename:
            line = ""

            with open(filename, 'r') as file:
                line = file.readline()

            line = f"{value} 0.500000 0.500000 1.000000 1.000000"

            with open(filename, 'w') as file:
                file.write(line)
```

```python
from ultralytics import YOLO
import torch
```

```python
torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Load the model
model = YOLO("yolov8s.pt")  # trying the pretrained small model version on 5 epochs

# Use the model
results = model.train(data="/content/drive/MyDrive/facial_expressions_dataset/hs_dataset/happy_sad.yaml",
            epochs=5,
            #resume = True,
            batch = -1,
            imgsz = 96,
            project = 'facial_expressions_dataset',
            #device='cpu'
            )
```
