```python
# creating new annotation files with the new class labels

img_train_path = '/content/drive/MyDrive/facial_expressions_dataset/only_four_emos/images/train'
img_val_path = '/content/drive/MyDrive/facial_expressions_dataset/only_four_emos/images/val'

label_train_path = '/content/drive/MyDrive/facial_expressions_dataset/only_four_emos/labels/train'
label_val_path = '/content/drive/MyDrive/facial_expressions_dataset/only_four_emos/labels/val'
```

Creating new annotation files for the smaller dataset with new class labels and moving them to the correct location in Google Drive

```python
def remove_nums(text):
    return ''.join(char for char in text if not char.isdigit())
```

```python
emo_list = ['anger', 'happy', 'sad', 'surprise']

#creating annotation files for val set

for i in os.listdir(img_val_path):
  f = open(f'{label_val_path}/{i[:-4]}.txt','w')
  emotion = remove_nums(i[:-4])
  f.write(f'{emo_list.index(emotion)} 0.500000 0.500000 1.000000 1.000000')
  f.close()
```

```python
#creating annotation files for train set

for i in os.listdir(img_train_path):
  f = open(f'{label_train_path}/{i[:-4]}.txt','w')
  emotion = remove_nums(i[:-4])
  f.write(f'{emo_list.index(emotion)} 0.500000 0.500000 1.000000 1.000000')
  f.close()
```
